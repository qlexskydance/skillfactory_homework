```python
def get_streets_address(address):
    exclude_list = ['N', 'W', 'S', 'E']
    address_list = address.split(' ')
    street_type = address_list[-1]
    if street_type in exclude_list:
        street_type = address_list[-2]
    return street_type
```

```python
street_types = melb_df['Address'].apply(get_streets_address)
display(street_types)
```

```python
print(street_types.nunique())
display(street_types.value_counts())
```

```python
popular_street = street_types.value_counts().nlargest(10).index
display(popular_street)
```

```python
melb_df['StreetType'] = street_types.apply(lambda x: x if x in popular_street else 'other')
display(melb_df['StreetType'].nunique())
```

```python
def get_experience(arg):
    month_key_words = ['месяц', 'месяцев', 'месяца']
    yers_key_words = ['год', 'лет', 'года']
    arg_splitted = arg.split(' ')
    month = 0
    year = 0
    for i in range(len(arg_splitted)):
        if arg_splitted[i] in month_key_words:
            month = arg_splitted[i-1]
        if arg_splitted[i] in yers_key_words:
            year = arg_splitted[i-1]
    return int(year)*12 + int(month) 
    
```
